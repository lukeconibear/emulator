#### Maxi-min Latin hypercube spacefilling design of simulator runs
#### For both training (50 runs) and testing (5 runs)
#### For each of the five features, designed over 100,000 iterations

```python
import pandas as pd
from pyDOE import lhs
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
design_train = lhs(5, samples=50, criterion='maximin', iterations=100000) * 1.5
design_test  = lhs(5, samples=5, criterion='maximin', iterations=100000) * 1.5
```

```python
df_train = pd.DataFrame(data=design_train, columns=['RES', 'IND', 'TRA', 'AGR', 'ENE'])
df_test  = pd.DataFrame(data=design_test, columns=['RES', 'IND', 'TRA', 'AGR', 'ENE'])
```

```python
df_train.to_csv('/nfs/see-fs-02_users/earlacoa/emulator/latin_hypercube_inputs_train.csv')
df_test.to_csv('/nfs/see-fs-02_users/earlacoa/emulator/latin_hypercube_inputs_test.csv')
```
